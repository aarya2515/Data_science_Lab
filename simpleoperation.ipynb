# Factorial

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Compute factorial for 4
number = 4
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

#Multiplication Table

```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

# Display table for 6
display_multiplication_table(6)
```

# odd&Even

```python
def check_odd_even(number):
    if number % 2 == 0:
        print(f"The number {number} is even.")
    else:
        print(f"The number {number} is odd.")

# Check if 8 is odd or even
check_odd_even(8)
```

# palindrome

```python
def is_palindrome(number):
    # Convert the number to a string
    num_str = str(number)
    # Reverse the string
    reversed_num_str = num_str[::-1]
    # Compare the original string with the reversed string
    return num_str == reversed_num_str

# Example usage:
# print(is_palindrome(121)) # Expected: True
# print(is_palindrome(123)) # Expected: False
```

**Reasoning**:
The previous step defined the `is_palindrome` function. The next step is to use this function to check if the number 7 is a palindrome, as requested by the overall task, and store the result.



```python
number_to_check = 7
is_seven_palindrome = is_palindrome(number_to_check)
print(f"Is the number {number_to_check} a palindrome? {is_seven_palindrome}")
```

## Define Decision Rules

### Subtask:
Establish a set of 'if-then' rules for a simple decision domain. For example, rules for advising on what to wear based on weather conditions (temperature and rain).


### Clothing Recommendation Decision Rules

Based on temperature and rain status, here are the established 'if-then' rules for clothing recommendations:

1.  **IF** Temperature is 'Hot' (>= 30°C) **AND** Rain Status is 'No Rain' **THEN** Recommend: 'Light clothing (shorts, t-shirt), hat, sunglasses'
2.  **IF** Temperature is 'Warm' (20-29°C) **AND** Rain Status is 'No Rain' **THEN** Recommend: 'T-shirt and shorts/light trousers'
3.  **IF** Temperature is 'Warm' (20-29°C) **AND** Rain Status is 'Light Rain' **THEN** Recommend: 'Light jacket or umbrella, light trousers'
4.  **IF** Temperature is 'Mild' (10-19°C) **AND** Rain Status is 'No Rain' **THEN** Recommend: 'Long-sleeved shirt, light jacket'
5.  **IF** Temperature is 'Mild' (10-19°C) **AND** Rain Status is 'Light Rain' **THEN** Recommend: 'Waterproof jacket, long-sleeved shirt'
6.  **IF** Temperature is 'Cold' (< 10°C) **AND** Rain Status is 'No Rain' **THEN** Recommend: 'Warm coat, scarf, gloves'
7.  **IF** Temperature is 'Cold' (< 10°C) **AND** Rain Status is 'Heavy Rain' **THEN** Recommend: 'Heavy waterproof coat, warm layers, waterproof boots, umbrella'

**Reasoning**:
The decision rules have been defined in a markdown cell. The next step is to implement these rules programmatically within a function that takes temperature and rain status as input and returns a clothing recommendation.



```python
def get_clothing_recommendation(temperature, rain_status):
    """
    Provides clothing recommendations based on temperature and rain status.

    Args:
        temperature (int): The current temperature in Celsius.
        rain_status (str): The current rain status ('No Rain', 'Light Rain', 'Heavy Rain').

    Returns:
        str: A clothing recommendation.
    """
    rain_status = rain_status.lower().replace(' ', '') # Normalize input

    if temperature >= 30:
        if rain_status == 'norain':
            return 'Light clothing (shorts, t-shirt), hat, sunglasses'
    elif 20 <= temperature <= 29:
        if rain_status == 'norain':
            return 'T-shirt and shorts/light trousers'
        elif rain_status == 'lightrain':
            return 'Light jacket or umbrella, light trousers'
    elif 10 <= temperature <= 19:
        if rain_status == 'norain':
            return 'Long-sleeved shirt, light jacket'
        elif rain_status == 'lightrain':
            return 'Waterproof jacket, long-sleeved shirt'
    elif temperature < 10:
        if rain_status == 'norain':
            return 'Warm coat, scarf, gloves'
        elif rain_status == 'heavyrain':
            return 'Heavy waterproof coat, warm layers, waterproof boots, umbrella'

    return 'No specific recommendation for these conditions.' # Default case if no rule matches

print("Defined function 'get_clothing_recommendation'.")
```

**Reasoning**:
The `get_clothing_recommendation` function has been defined. To demonstrate its usage and fulfill the task of presenting an example, I will now call this function with sample temperature and rain status values and print the returned recommendation.



```python
current_temperature = 25
current_rain_status = 'No Rain'
clothing_recommendation = get_clothing_recommendation(current_temperature, current_rain_status)
print(f"For a temperature of {current_temperature}°C and '{current_rain_status}', the recommendation is: {clothing_recommendation}")

current_temperature = 15
current_rain_status = 'Light Rain'
clothing_recommendation = get_clothing_recommendation(current_temperature, current_rain_status)
print(f"For a temperature of {current_temperature}°C and '{current_rain_status}', the recommendation is: {clothing_recommendation}")
```

## Create User Interface

### Subtask:
Create a simple text-based interface to allow the user to input information and receive the expert system's decision.


```python
print("\n--- Clothing Recommendation Expert System ---")

while True:
    try:
        user_temperature = int(input("Enter the current temperature in Celsius (e.g., 25): "))
        break
    except ValueError:
        print("Invalid input. Please enter an integer for temperature.")

valid_rain_statuses = ['No Rain', 'Light Rain', 'Heavy Rain']
while True:
    user_rain_status = input(f"Enter the rain status ({', '.join(valid_rain_statuses)}): ")
    if user_rain_status in valid_rain_statuses:
        break
    else:
        print("Invalid rain status. Please choose from the provided options.")

final_recommendation = get_clothing_recommendation(user_temperature, user_rain_status)
print(f"\nBased on your input (Temperature: {user_temperature}°C, Rain Status: '{user_rain_status}'), the recommendation is: {final_recommendation}")
```
